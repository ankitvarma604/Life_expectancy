```python
z = ['infant_deaths','under-five_deaths','Measles']
x = df[z]
y = df['Population']
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
std =StandardScaler()
x_s = std.fit_transform(x)
x_s = pd.DataFrame(x_s, columns=z)
x_s
```

```python
wcss =[]
for i in range (1,15):
    kmean = KMeans(n_clusters=i,random_state=42)
    kmean.fit(x_s)
    wcss.append(kmean.inertia_)
plt.plot(range(1,15),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS');
```

```python
kmean = KMeans(n_clusters = 4)
y_val = kmean.fit_predict(x_s)
len(y_val)
```

```python
y = pd.DataFrame(y)
```

```python
y['cluseter'] = y_val
```

```python
x = [0,1,2,3]
for i in x:
    val = y[y['cluseter'] == x[i]]
    z = round(val['Population'].mean(),2)
    print("for cluser {} mean age is {}".format(x[i],z))
```

```python
def mean_age(col):
    pop = col[0]
    cluster = col[1]
    if pd.isnull(pop):
        if cluster == 0:
            return 8239232.46
        if cluster == 1:
            return 449472510.47
        if cluster == 2:
            return 46941420.17
        if cluster == 3:
            return 28908041.43
    return pop    
```

```python
y['Population'] = y[['Population','cluseter']].apply(mean_age, axis = 1)
```

```python
kmean = KMeans(n_clusters = 4)
y_val = kmean.fit_predict(x_s)
y = pd.DataFrame(y)
y['cluster'] = y_val
x = [0,1,2,3]
for i in x:
    val = y[y['cluster'] == x[i]]
    z = round(val['GDP'].mean(),2)
    print("for cluser {} mean GDP is {}".format(x[i],z))
```

```python
def mean_gdp(col):
    gdp = col[0]
    cluster = col[1]
    if pd.isnull(gdp):
        if cluster == 0:
            return 1027.15
        if cluster == 1:
            return 2934.21
        if cluster == 2:
            return 50779.75
        if cluster == 3:
            return 12338.44
    return gdp    
```

```python
y['GDP'] = y[['GDP','cluster']].apply(mean_gdp, axis = 1)
df['GDP']= y['GDP']
```
