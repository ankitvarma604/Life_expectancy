```python
def mean_age(col):
    pop = col[0]
    cluster = col[1]
    if pd.isnull(pop):
        if cluster == 0:
            return 8239232.46
        if cluster == 1:
            return 449472510.47
        if cluster == 2:
            return 46941420.17
        if cluster == 3:
            return 28908041.43
    return pop    
```

```python
y['Population'] = y[['Population','cluseter']].apply(mean_age, axis = 1)
```

```python
kmean = KMeans(n_clusters = 4)
y_val = kmean.fit_predict(x_s)
y = pd.DataFrame(y)
y['cluster'] = y_val
x = [0,1,2,3]
for i in x:
    val = y[y['cluster'] == x[i]]
    z = round(val['GDP'].mean(),2)
    print("for cluser {} mean GDP is {}".format(x[i],z))
```

```python
def mean_gdp(col):
    gdp = col[0]
    cluster = col[1]
    if pd.isnull(gdp):
        if cluster == 0:
            return 1027.15
        if cluster == 1:
            return 2934.21
        if cluster == 2:
            return 50779.75
        if cluster == 3:
            return 12338.44
    return gdp    
```

```python
y['GDP'] = y[['GDP','cluster']].apply(mean_gdp, axis = 1)
df['GDP']= y['GDP']
```

### Dealing with Outliers


```python
def draw_fig(df,val):
    plt.figure(figsize=(25,8), facecolor='white')
    plt.subplot(1,2,1)
    sns.distplot(df[val])
    plt.subplot(1,2,2)
    sns.boxplot(df[val]);
```

```python
draw_fig(df,'Life_expectancy')
```

```python
def find_oulier_limits(df,val):
    q1,q2 = np.percentile(df[val],[25,75])
    print("Q1 and Q2 are {} and {}".format(q1,q2))
    iqr =q2-q1
    u_limit = q2+(1.5*iqr)
    l_limit = q1-(1.5*iqr)
    print("Lower limit is {} and Upper limit is {}".format(round(l_limit,2),round(u_limit,2)))
```

```python
find_oulier_limits(df,'Life_expectancy')
```
