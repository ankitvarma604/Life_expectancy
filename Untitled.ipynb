```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("performance_by_publisher_24-may-2021_-_31-may-2021_daily.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
pub = df['Publisher Name'].unique().tolist()
```

```python
len(pub)
```

```python

for i in range(len(pub)):
    v = df[df['Publisher Name'] == pub[i]].reset_index(drop =True)
    v['Period'] = pd.to_datetime(v['Period']).dt.date
    v = v.sort_values('Period',ascending = True)
    sns.barplot(x = v['Period'], y = v['Sale Amount (USD)'])
    plt.xlabel(v['Publisher Name'][0]) 
```
